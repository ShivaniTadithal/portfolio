```python
import pandas as pd
from tabulate import tabulate

# Data for each year
data_2017 = {
    'Year': [2017],
    'Stock Price': [32.65],
    'EPS': [2.37],
    'Book Value per Share': [5.90],
    'Sales per Share': [22.47],
    'Dividends per Share': [2.80],
}

data_2018 = {
    'Year': [2018],
    'Stock Price': [39.30],
    'EPS': [1.70],
    'Book Value per Share': [6.30],
    'Sales per Share': [23.75],
    'Dividends per Share': [2.80],
}

data_2019 = {
    'Year': [2019],
    'Stock Price': [49.96],
    'EPS': [2.98],
    'Book Value per Share': [7.20],
    'Sales per Share': [24.38],
    'Dividends per Share': [2.80],
}

# Create DataFrames for each year
df_2017 = pd.DataFrame(data_2017)
df_2018 = pd.DataFrame(data_2018)
df_2019 = pd.DataFrame(data_2019)

# Calculate valuation ratios for each DataFrame
for df in [df_2017, df_2018, df_2019]:
    df['P/E Ratio'] = df['Stock Price'] / df['EPS']
    df['P/B Ratio'] = df['Stock Price'] / df['Book Value per Share']
    df['P/S Ratio'] = df['Stock Price'] / df['Sales per Share']
    df['Dividend Yield'] = df['Dividends per Share'] / df['Stock Price'] * 100

# Print the DataFrames as tables using the tabulate library
print("2017\n" + tabulate(df_2017, headers='keys', tablefmt='pretty'))
print("\n2018\n" + tabulate(df_2018, headers='keys', tablefmt='pretty'))
print("\n2019\n" + tabulate(df_2019, headers='keys', tablefmt='pretty'))
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2017, 2018, 2019]

# Collect the valuation metrics for each year
pe_ratios = [df_2017['P/E Ratio'][0], df_2018['P/E Ratio'][0], df_2019['P/E Ratio'][0]]
pb_ratios = [df_2017['P/B Ratio'][0], df_2018['P/B Ratio'][0], df_2019['P/B Ratio'][0]]
ps_ratios = [df_2017['P/S Ratio'][0], df_2018['P/S Ratio'][0], df_2019['P/S Ratio'][0]]
dividend_yields = [df_2017['Dividend Yield'][0], df_2018['Dividend Yield'][0], df_2019['Dividend Yield'][0]]

# Define a function to create a bar chart for each metric
def create_bar_chart(metric_values, metric_name, ylabel):
    bar_width = 0.5
    plt.figure(figsize=(10, 5))
    plt.bar(years, metric_values, width=bar_width, color='b', alpha=0.7)
    plt.xlabel('Years')
    plt.xticks(years)
    plt.ylabel(ylabel)
    plt.title(f'{metric_name} (2017-2019)')
    plt.show()

# Create bar charts for each valuation metric
create_bar_chart(pe_ratios, 'P/E Ratio', 'Ratio')
create_bar_chart(pb_ratios, 'P/B Ratio', 'Ratio')
create_bar_chart(ps_ratios, 'P/S Ratio', 'Ratio')
create_bar_chart(dividend_yields, 'Dividend Yield', 'Percentage')
```

```python
data_2017_gsk = {'P/E Ratio': [41.50], 'P/B Ratio': [2.34], 'P/S Ratio': [3.19], 'Dividend Yield': [6.14]}
df_2017_gsk = pd.DataFrame(data_2017_gsk)

data_2018_gsk = {'P/E Ratio': [20.79], 'P/B Ratio': [2.43], 'P/S Ratio': [3.49], 'Dividend Yield': [5.34]}
df_2018_gsk = pd.DataFrame(data_2018_gsk)

data_2019_gsk = {'P/E Ratio': [15.86], 'P/B Ratio': [2.49], 'P/S Ratio': [3.93], 'Dividend Yield': [4.37]}
df_2019_gsk = pd.DataFrame(data_2019_gsk)

def create_comparison_bar_chart(metric_values_az, metric_values_gsk, metric_name, ylabel):
    bar_width = 0.35
    plt.figure(figsize=(10, 5))
    
    x = np.arange(len(years))
    
    plt.bar(x - bar_width/2, metric_values_az, width=bar_width, color='b', alpha=0.7, label='AstraZeneca')
    plt.bar(x + bar_width/2, metric_values_gsk, width=bar_width, color='r', alpha=0.7, label='GlaxoSmithKline')
    
    plt.xlabel('Years')
    plt.xticks(x, years)
    plt.ylabel(ylabel)
    plt.title(f'{metric_name} Comparison (2017-2019)')
    plt.legend()
    plt.show()

# Collect the valuation metrics for each year for AstraZeneca and GlaxoSmithKline
years = ['2017', '2018', '2019']

pe_ratios_az = [df_2017['P/E Ratio'][0], df_2018['P/E Ratio'][0], df_2019['P/E Ratio'][0]]
pe_ratios_gsk = [df_2017_gsk['P/E Ratio'][0], df_2018_gsk['P/E Ratio'][0], df_2019_gsk['P/E Ratio'][0]]

pb_ratios_az = [df_2017['P/B Ratio'][0], df_2018['P/B Ratio'][0], df_2019['P/B Ratio'][0]]
pb_ratios_gsk = [df_2017_gsk['P/B Ratio'][0], df_2018_gsk['P/B Ratio'][0], df_2019_gsk['P/B Ratio'][0]]

ps_ratios_az = [df_2017['P/S Ratio'][0], df_2018['P/S Ratio'][0], df_2019['P/S Ratio'][0]]
ps_ratios_gsk = [df_2017_gsk['P/S Ratio'][0], df_2018_gsk['P/S Ratio'][0], df_2019_gsk['P/S Ratio'][0]]

dividend_yields_az = [df_2017['Dividend Yield'][0], df_2018['Dividend Yield'][0], df_2019['Dividend Yield'][0]]
dividend_yields_gsk = [df_2017_gsk['Dividend Yield'][0], df_2018_gsk['Dividend Yield'][0], df_2019_gsk['Dividend Yield'][0]]

# Create comparison bar charts
create_comparison_bar_chart(pe_ratios_az, pe_ratios_gsk, 'P/E Ratio', 'P/E Ratio')
create_comparison_bar_chart(pb_ratios_az, pb_ratios_gsk, 'P/B Ratio', 'P/B Ratio')
create_comparison_bar_chart(ps_ratios_az, ps_ratios_gsk, 'P/S Ratio', 'P/S Ratio')
create_comparison_bar_chart(dividend_yields_az, dividend_yields_gsk, 'Dividend Yield', 'Dividend Yield (%)')
```
